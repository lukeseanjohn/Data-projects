Below I will try to make other correlations as to what is influencing the insurance charges

```python
#Showing age and gender demographics may help visualise the data
plt.figure(figsize=(8, 5))
sns.kdeplot(data=insurance_df, x="age", hue="sex", multiple="stack")
plt.title('Distribution of age by sex')
plt.xlabel('Age')
plt.ylabel('density')
plt.show()
```

```python
# [Analysiis of above graph showing age and sex demographics]
This shows a greater majority of the group are female and a greater majority of the group are towards 20 -30 years old and also a higher number between the age of 50 - 60. This suggests times of life where people are more likely to buy health insurance.
```

```python
# [Displays distribution of charges by age more clearly and concisely] 
plt.figure(figsize=(10, 6))
sns.kdeplot(data=insurance_df, x="charges", hue="age", multiple="stack")
plt.title('Distribution of charges by age')
plt.xlabel('charges')
plt.ylabel('Density')
plt.show()
```

```python
# [Analysis on the kde plot]
This graph presents the correlation between age and charges more specifically. The data supports the hypothesis that there is a gradual increase in charges as people get older. 
```

 A violin plot 

```python
# [Creating a violin plot to present the data based on smokers, non - smokers and region]
plt.figure(figsize=(10, 6))
sns.violinplot(x="region", y="charges", data=insurance_df, hue="smoker")
plt.title('Insurance Charges by Smoking Status')
plt.xlabel('Smoker')
plt.ylabel('Charges')
plt.show()
```

```python
# [Analysis on the violin plot]

This graph shows that on average smokers pay more than non - smokers. However there seems to be a larger ditribution in the 'smoker' demographic suggesting other considerations that influence the insurance charges for smokers but less so for non - smokers. There doesn't seem to a huge disparity in the influence that region has on insurance charges. It seems southeast has a small number of the highest charges. The northeast charges seem to be more grouped towards the lower end of the spectrum. 
```

```python
#[Creating a multi dimensional data visulisation graph to further understand bmi and children influences]
sns.pairplot(insurance_df, hue="smoker", diag_kind="kde")
plt.suptitle('Scatter Plot Matrix of Insurance Data', y=1.02)
plt.show()
```
